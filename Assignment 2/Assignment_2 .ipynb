```python
# Rejection Method
points_in_circle = []
points_out_circle = []

# Generate points
while len(points_in_circle) <= N:
    x = np.random.uniform(-R, R)
    y = np.random.uniform(-R, R)
    point = np.array([x, y])

    # Check if the point is inside the circle
    if x**2 + y**2 <= R**2:
        points_in_circle.append(point)
    else:
        points_out_circle.append(point)

# Convert to NumPy arrays
points_in_circle = np.array(points_in_circle)
points_out_circle = np.array(points_out_circle)

# Plot the points
plt.figure(figsize=[7, 7])
plt.scatter(points_in_circle[:, 0], points_in_circle[:, 1], s=3, edgecolors='b', label='Inside Circle')
plt.scatter(points_out_circle[:, 0], points_out_circle[:, 1], s=3, edgecolors='r', label='Outside Circle')
plt.xlabel('x', fontsize=10)
plt.ylabel('y', fontsize=10)
plt.title("Rejection Method", fontsize=12)
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

The reason the distribution of points is not uniform over the circle of radius R when generating points uniformly with respect to the radius (r) is because the area of a circular annulus (a ring-shaped region) at different radii is not the same. Points generated uniformly with respect to the radius will naturally result in more points being concentrated at smaller radii, and fewer points at larger radii.

Here's a simple way to understand this:

- If you generate points uniformly with respect to the radius (r), it means that you are equally likely to select any value of r within the range [0, R], where R is the maximum radius of the circle.

- Consider two concentric circular annuli (r to r and r to r, where r < r < r) with the same width. The area of the inner annulus (r to r) is smaller than the area of the outer annulus (r to r).

- When you generate points uniformly with respect to the radius, you have an equal chance of selecting any radius within [r, r] and [r, r]. Since the area of the outer annulus is larger, you are more likely to select a point in the outer annulus, leading to a concentration of points at larger radii.

This non-uniformity in the distribution of points can be problematic if you want a truly uniform distribution of points on the circle's circumference. To achieve a uniform distribution on the circumference, you should sample angles uniformly (as described in my previous response) because points on the circumference of a circle are determined by their angles, not their radii.

### c)

Here, we will compute the value of $C$ using the given distribution of $r$, which is defined as $p(r) = C2\pi r$. The calculation proceeds as follows:

Starting with the cumulative distribution function (CDF):

$$
F(r) = \int_{0}^{r} C 2 \pi x\ dx
$$

Integrating with respect to $x$:

$$
F(r) = C2\pi \int_{0}^{r}x\ dx = C 2\pi \frac{r^2}{2} = C\pi r^2
$$

To find the constant $C$, we set $F(r = R) = 1$, which gives us:

$$
C \pi R^2 = 1
$$

Solving for $C$:

$$
C = \frac{1}{R^2 \pi}\ \Big|_{R = 1} = \frac{1}{\pi}
$$

Now that we have the constant C computed as \(\frac{1}{\pi}\), we can calculate the inverse of the cumulative distribution as:

$$
\begin{aligned}
F(r) &= C \pi r^2 = y \\
r^2 &= \frac{y}{C \pi} = \frac{y}{1} = y \\
r &= \sqrt{y}
\end{aligned}
$$

This simplifies the calculation process, making it clearer and more concise.


```python
# Calculate the normalization constant
C = 1 / (np.pi * R**2)

# Generate random points using the inverse distribution method
theta = np.random.uniform(0, 2 * np.pi, N)
r = np.sqrt(np.random.uniform(0, 1, N))

# Convert to Cartesian coordinates
x = r * np.cos(theta)
y = r * np.sin(theta)

# Create the scatter plot
plt.figure(figsize=[7, 7])
plt.scatter(x, y, s=1, facecolors='none', edgecolors='b')
plt.xlabel('x', fontsize=10)
plt.ylabel('y', fontsize=10)
plt.title("Inverse Distribution Method", fontsize=12)
plt.grid()
plt.tight_layout()
plt.show()
```

### a)

**1. Finding the constant C:**

We know that the density function of the random variable x is given by:
$$f(x) = Cx^{-\frac{3}{2}}$$

To find the value of $C$, we need to integrate this function over its support $[1, \infty)$ and set it equal to 1:

$$
\begin{align*}
\int_{1}^{\infty} Cx^{-\frac{3}{2}}\ dx &= 1 \\
C \int_{1}^{\infty} x^{-\frac{3}{2}}\ dx &= 1 \\
C \left[\frac{x^{-\frac{1}{2}}}{-\frac{1}{2}}\right]_{1}^{\infty} &= 1 \\
-2C \left[x^{-\frac{1}{2}}\right]_{1}^{\infty} &= 1 \\
-2C \left(0 - \frac{1}{1}\right) &= 1 \\
2C &= 1 \\
C &= \frac{1}{2}
\end{align*}
$$

So, we have found that $C = \frac{1}{2}$.

**2. Finding the cumulative distribution function (CDF):**

Now that we have the value of C, we can find the cumulative distribution function (CDF) as follows:

$$
\begin{align*}
F(x) &= \int_{1}^{x} \frac{1}{2} r^{-\frac{3}{2}}\ dr \\
&= \frac{1}{2} \int_{1}^{x} r^{-\frac{3}{2}}\ dr \\
&= \frac{1}{2} \left[\frac{r^{-\frac{1}{2}}}{-\frac{1}{2}}\right]_{1}^{x} \\
&= \left[-r^{-\frac{1}{2}}\right]_{1}^{x} \\
&= -x^{-\frac{1}{2}} + 1
\end{align*}
$$

So, the cumulative distribution function (CDF) is $F(x) = 1 - x^{-\frac{1}{2}}$.

**3. Finding the inverse CDF:**

To find the inverse of the CDF, we set $F(x) = y$ and solve for x:

$$
\begin{align*}
1 - x^{-\frac{1}{2}} &= y \\
1 - y &= x^{-\frac{1}{2}} \\
\frac{1}{(1 - y)^2} &= x
\end{align*}
$$

Thus, the inverse of the CDF is $F^{-1}(y) = \frac{1}{(1 - y)^2}$.


```python
def inv_CDF(y, A):
    y_scaled = y/A
    return 1/ np.square(1-y_scaled)
```

- Slower method.
